```python
def sstrip(text):
    try:
        return text.strip()
    except AttributeError:
        return text
def make_int(text):
    return int(text.strip('"'))

table = pd.read_table('NYCPark_Crime_Stat_Tidy.csv', sep=r',', 
                     converters = {'BOROUGH':sstrip,
                                  'PARK': sstrip})
table.BOROUGH.unique()
```

```python
# Removing Mis Spelled Borough Names
table.loc[table['BOROUGH']=='BR0NX', 'BOROUGH']='BRONX'
table.loc[table['BOROUGH']=='QNS', 'BOROUGH']='QUEENS'
table.loc[table['BOROUGH']=='queens', 'BOROUGH']='QUEENS'
table.loc[table['BOROUGH']=='BROOOKLYN', 'BOROUGH']='BROOKLYN'
table.loc[table['BOROUGH']=='BROKLYN', 'BOROUGH']='BROOKLYN'
table.to_csv('NYCPark_Crime_Stat_Tidy.csv', index=False)
CLEAN_DATA = pd.read_csv('NYCPark_Crime_Stat_Tidy.csv')
CLEAN_DATA.head(10)
```
